```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.metrics import classification_report
```

```python
# Given DataFrames
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Read the 2019 training csv file into a pandas dataFrame
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))

# inspect the training DataFrame
pd.set_option('display.max_columns', 86)
test_df.head()
```

```python
# confirm inspection of training dataframe
test_df.dtypes
```

```python
# drop rows with missing data
test_df = test_df.dropna()

# drop columns with missing data
test_df = test_df.dropna(axis='columns')

# the 'unnamed 0'index 0 and the name 'index' columns appear to be the same, one needs to go
test_df = test_df.drop(columns=["Unnamed: 0"])

# column names
test_df.columns
```

```python
# Identify all the possible values in the loan_status column
list(test_df['loan_status'].unique())
```

```python
#Now clean, read the loan_status column dataset
X = test_df.drop('loan_status', axis=1)
X
```

```python
# Convert output labels to 0 and 1
y_label = LabelEncoder().fit_transform(test_df['loan_status'])
y_label
```

```python
# One-hot encoding the entire 2020 dataframe
X_dummies = pd.get_dummies(X)
print(X_dummies.columns)
X_dummies
```

```python
# check 2020 columns
len(X_dummies.columns)
```

```python
y = test_df['loan_status']
target_names = ['low_risk', 'high_risk']
```

```python
# Prepare the 2020 test data
X_train, X_test, y_train, y_test = train_test_split(X_dummies, y_label, random_state=1)
```

```python
# Create a logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
# Fit (train) the 2019 model by using the training data
classifier.fit(X_train, y_train)
```

```python
# Validate the model by using the test data
print(f"Training Logistic Regression Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Logistic Regression Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_true = y_test
y_pred = classifier.predict(X_test)
confusion_matrix(y_true, y_pred)
```

```python
confusion_matrix(y, classifier.predict(X))
```

```python
print(classification_report(y_true, y_pred))
```
