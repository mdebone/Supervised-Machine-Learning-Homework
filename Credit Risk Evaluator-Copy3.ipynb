```python
# encode string values to numeric values
debt_flag_dict = {'N': 0, 'Y': 1}
trained_df7 = trained_df6.replace({'debt_settlement_flag': debt_flag_dict})
trained_df7.head()
```

```python
# Drop the loan_status to create the X data
X = trained_df7.drop('loan_status', axis=1)
y = trained_df7['loan_status']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
# Fit (train) our model by using the training data
classifier.fit(X_train, y_train)
```

```python
# Validate the model by using the test data
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_true = y_test
y_pred = classifier.predict(X_test)
confusion_matrix(y_true, y_pred)
```

```python
confusion_matrix(y, classifier.predict(X))
```
